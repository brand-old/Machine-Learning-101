```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection

import pandas as pd
import numpy as np

train_file = r"C:\Data Analytics\Machine Learning\Practice\Logistic Regression Dataset\0000000000002429_training_titanic_x_y_train.csv"
test_file = r"C:\Data Analytics\Machine Learning\Practice\Logistic Regression Dataset\0000000000002429_test_titanic_x_test.csv"
#train_file = "0000000000002429_training_titanic_x_y_train.csv"
#test_file = "0000000000002429_test_titanic_x_test.csv"

train = np.genfromtxt(train_file, delimiter=",", skip_header=1)
x_tr = train[:,:-1]
    #.reshape(-1,1)
y_train = train[:,-1]
test = np.genfromtxt(test_file, delimiter=",", skip_header=1)
x_te = test
    #.reshape(-1,1)
#y_test =data2[:,-1]

#x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.3, random_state=42)

x_train_cleaned = pd.DataFrame(x_tr).dropna(axis=1, how='all')
#same columns from test dataset
x_test_cleaned = pd.DataFrame(x_te)[x_train_cleaned.columns]

# data cleaning
from sklearn.impute import SimpleImputer
nan_handler = SimpleImputer(strategy='mean')

x_train = nan_handler.fit_transform(x_train_cleaned)
x_test = nan_handler.transform(x_test_cleaned)

print(x_train.shape)
print(x_test.shape)

print(x_train)
print(x_test)

class_ = LogisticRegression( C = 1.5, max_iter = 1000, solver = 'liblinear')
class_.fit(x_train, y_train)

y_pred = class_.predict(x_test)

score_train = class_.score(x_train, y_train)
#score_test = class_.score(x_test, y_test)

#class_.predict(cancer.data) - cancer.target
#class_.predict_proba(cancer.data)

print(round(score_train, 3))
#print(f"{score_test:.3f}")

print(y_pred)

np.savetxt(r"C:\Data Analytics\Machine Learning\Practice\Logistic Regression Dataset\Titanic_Prediction.csv", X = y_pred, fmt='%d', delimiter=",")
```
