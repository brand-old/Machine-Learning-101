```python
import numpy as np
import os

train_file = r"C:\Data Analytics\Machine Learning\Linear Regression Diabetes Dataset\diabetes_x_y_train.csv"
test_file = r"C:\Data Analytics\Machine Learning\Linear Regression Diabetes Dataset\diabetes_x_test.csv"
#train_file = "diabetes_x_y_train.csv"
#test_file = "diabetes_x_test.csv"

data = np.genfromtxt(train_file, delimiter=",", skip_header=1)
x_train = data[:,:-1]
    #.reshape(-1,1)
y_train = data[:,-1]
data2 = np.genfromtxt(test_file, delimiter=",")
x_test = data2
    #.reshape(-1,1)
#y_test =data2[:,-1]

#print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)

#from sklearn import model_selection
#x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, test_size = 0.3, random_state = 42)
#print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

from sklearn.linear_model import LinearRegression
algo1 = LinearRegression()
algo1.fit(x_train, y_train)

y_pred = algo1.predict(x_test)

#print(np.round(algo1.coef_[0], 2))
#print(np.round(algo1.intercept_, 2))

score_training = algo1.score(x_train, y_train)
#score_test = algo1.score(x_test, y_pred)


print("Train Score:", round(score_training, 5))
#print(f"Test Score: {score_test:.5f}")

np.savetxt(r"C:\Data Analytics\Machine Learning\Linear Regression Diabetes Dataset\Prediction.csv", X = y_pred, fmt='%.5f', delimiter=",")
```
