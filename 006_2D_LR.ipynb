```python
#import pandas as pd
#import numpy as np
#from sklearn.model_selection import train_test_split
#from sklearn.linear_model import LinearRegression

#df = pd.read_csv("datase.csv")

from sklearn import datasets
sugar = datasets.load_diabetes()

x = sugar.data
y = sugar.target
#x.shape

import pandas as pd
frame = pd.DataFrame(x)

#print(sugar.feature_names)

frame.columns = sugar.feature_names
frame['age_age'] = frame['age']**2

x2 = frame.values

from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, random_state=69)
x2_train, x2_test, y2_train, y2_test = model_selection.train_test_split(x2,y, random_state=69)

from sklearn.linear_model import LinearRegression
algo1 = LinearRegression()
algo2 = LinearRegression()

algo1.fit(x_train, y_train)
algo2.fit(x2_train, y2_train)

y_pred = algo1.predict(x_test)
y2_pred = algo2.predict(x2_test)

train_score = algo1.score(x_train, y_train)
test_score = algo1.score(x_test, y_test)

train2_score = algo2.score(x2_train, y2_train)
test2_score = algo2.score(x2_test, y2_test)

print("Train score:", train_score)
print("Train2 score:", train2_score)

print("Test score:", test_score)
print("Test2 score:", test2_score)

if test2_score > test_score:
    print("Score Improved")
else:
    print("score not improved")
```
